### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

x = list(range(11))
y = [i**2 for i in x]

plt.plot(x, y)
plt.title("Line Plot of y = xÂ²")
plt.xlabel("X values")
plt.ylabel("Y values (x squared)")
plt.grid(True)
plt.show()
```
